```python
from sqlalchemy import types

engine = easyConnect()

temp = pd.read_sql('temp_joiner', con=engine)

tt = temp.pivot(index=['ts_code','trade_date'],
                columns='data_name',
                values='data_value').reset_index()

# 明确类型，避免被映射成 TEXT
tt['ts_code'] = tt['ts_code'].astype(str)
tt['trade_date'] = pd.to_datetime(tt['trade_date']).dt.date  # or keep as string

tt.to_sql(
    'temp_joiner_wide',
    con=engine,
    if_exists='replace',
    index=False,   # 关键：不要把 DataFrame 的索引写到库里
    dtype={
        'ts_code': types.VARCHAR(12),      # 根据你的代码长度调整
        'trade_date': types.DATE           # 或 types.VARCHAR(10)
    },
    chunksize=50_000,   # 你有 ~75万行，分批更稳
    method='multi'
)
```
